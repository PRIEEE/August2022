```python
# import pandas
import pandas as pd
```

```python
# load some sample data
df = pd.read_excel('https://s3.eu-west-1.amazonaws.com/neueda.conygre.com/pydata/salesfunnel.xlsx')

df.head()
```

```python
# create a pivot table by Rep
pd.pivot_table(df,index='Rep')
```

```python
# create a pivot table on multiple index and values with multiple aggregates
import numpy as np

pivot_df = pd.pivot_table(
    df,index=['Manager','Rep','Product'],values=['Account','Price','Quantity'],
    aggfunc={
        'Account':np.count_nonzero,
        'Price':np.sum,
        'Quantity':[np.min,np.max]
    }
)
pivot_df
```
