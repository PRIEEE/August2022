```python
# import pandas and numpy
import pandas as pd
import numpy as np
```

```python
# Read data into a DataFrame
df_SPX = pd.read_csv('https://s3.eu-west-1.amazonaws.com/neueda.conygre.com/pydata/SPX.csv',
                     index_col='Date', parse_dates=True)

# Use the dtypes attribute to check what types are in each column
# the word 'object' is used to denote a string
print(df_SPX.dtypes)

df_SPX.head()
```

```python
# Convert the value in the 'Price' column from a String to a numeric (notice we also strip out the ",")
# Now check the dtypes and compare to the previous cell - price is now a "float64" i.e. a decimal number
df_SPX['Price'] = pd.to_numeric(df_SPX['Price'].str.replace(',',''))
df_SPX['Open'] = pd.to_numeric(df_SPX['Open'].str.replace(',',''))
df_SPX['High'] = pd.to_numeric(df_SPX['High'].str.replace(',',''))
df_SPX['Low'] = pd.to_numeric(df_SPX['Low'].str.replace(',',''))
df_SPX['Change %'].update
df_SPX['Change %'] = pd.to_numeric(df_SPX['Change %'].str.replace('%',''))
print(df_SPX.dtypes)

df_SPX.head()
```

```python
date = df_SPX.sort_index(axis = 0,ascending = True)
display(date)
```

```python
# include %matplotlib inline
%matplotlib inline

cols=['High','Low']
date = date.query("'2015-06-01'<Date and Date<'2016-06-30'")
```
